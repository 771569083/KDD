```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("DBSCAN_TASK1.csv")
df.head()
X = df.iloc[:,:].values.reshape(-1,2)
plt.figure(figsize=(16,9))
plt.plot(X[:,0],X[:,1],"bo")
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.079,min_samples=13).fit(X)
y_pred = dbscan.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```
