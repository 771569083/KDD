```python
# -*- coding: UTF-8 -*-
# @Author  : Wang Huanchen <11810419@mail.sustech.edu.cn>
# @Time    : 2021/3/13 10:15
import pandas as pd
import chardet
import os

with open("HW1data.csv", 'rb') as rawData:
    result = chardet.detect(rawData.read(os.path.getsize('HW1data.csv') // 100))
HW1 = pd.read_csv('HW1data.csv', encoding=result['encoding'])
temp = HW1[['PatientId', 'Gender', 'Age', 'Neighbourhood']]
temp = temp.dropna(subset=['PatientId'])
temp = temp.drop_duplicates()
id_ls = []
gender_ls = []
age_ls = []
neiB_ls = []
dic = {}
for row in temp.itertuples():
    Id = getattr(row, 'PatientId')
    dic[Id] = {'Gender': 'nan', 'Age': 'nan', 'Neighbourhood': 'nan'}
for row in temp.itertuples():
    id = getattr(row, 'PatientId')
    if id not in id_ls:
        gender = getattr(row, 'Gender')
        age = getattr(row, 'Age')
        neiB = getattr(row, 'Neighbourhood')
        if dic[id]['Gender'] == 'nan' and str(gender) != 'nan':
            dic[id]['Gender'] = str(gender)
        if dic[id]['Age'] == 'nan' and str(age) != 'nan':
            dic[id]['Age'] = age
        if dic[id]['Neighbourhood'] == 'nan' and str(neiB) != 'nan':
            dic[id]['Neighbourhood'] = str(neiB)
        if dic[id]['Gender'] != 'nan' and dic[id]['Age'] != 'nan' and dic[id]['Neighbourhood'] != 'nan':
            id_ls.append(id)
            gender_ls.append(dic[id]['Gender'])
            age_ls.append(dic[id]['Age'])
            neiB_ls.append(dic[id]['Neighbourhood'])
d = {'PatientId': id_ls, 'Gender': gender_ls, 'Age': age_ls, 'Neighbourhood': neiB_ls}
df = pd.DataFrame(data=d)
# To above, which is the step 2.a
HW1 = HW1.dropna(subset=['PatientId', 'ScheduledDay', 'AppointmentDay', 'SMS_received', 'No-show'])
# To above, which is the step 2.b
HW1 = HW1.merge(df, how='left', on='PatientId')
HW1 = HW1.rename(columns={'Gender_x': 'Gender', 'Age_x': 'Age', 'Neighbourhood_x': 'Neighbourhood'})
HW1['Gender'] = HW1['Gender_y']
HW1['Age'] = HW1['Age_y']
HW1['Neighbourhood'] = HW1['Neighbourhood_y']
HW1 = HW1.drop(columns=['Gender_y', 'Age_y', 'Neighbourhood_y'])
# To above, which is the step 2.c
HW1 = HW1.dropna(
    subset=['PatientId', 'Gender', 'Age', 'Neighbourhood', 'ScheduledDay', 'AppointmentDay', 'SMS_received', 'No-show'])
# To above, which is the step 3.
values = {'Scholarship': 0, 'Hipertension': 0, 'Diabetes': 0, 'Alcoholism': 0, 'Handcap': 0}
HW1 = HW1.fillna(value=values)
# To above, which is the step 4.
HW1 = HW1.drop(columns=['PatientId', 'AppointmentID'])
# To above, which is the step 5.
HW1 = HW1[HW1['Age'] >= 0]
# To above, which is the step 6.
HW1['ScheduledDay'] = pd.to_datetime(HW1['ScheduledDay'])
HW1['AppointmentDay'] = pd.to_datetime(HW1['AppointmentDay'])
HW1['Delta_Day'] = (HW1['AppointmentDay'] - HW1['ScheduledDay']).dt.days
HW1 = HW1[HW1['Delta_Day'] >= 0]
# To above, which is the step 7.
HW1['SDay_DOW'] = HW1['ScheduledDay'].dt.weekday + 1
HW1['ADay_DOW'] = HW1['AppointmentDay'].dt.weekday + 1
# To above, which is the step 8.
HW1 = HW1.drop(columns=['ScheduledDay', 'AppointmentDay'])
# To above, which is the step 9.
```
